```python
import os
import csv
from langdetect import detect
# Specify the path to the main 'year' folder
year_folder_path = "2023"  # Replace with your actual path

# Specify the path to the CSV file where you want to store the data
csv_file_path = "dataset2023.csv"  # Replace with your desired CSV file path

# Create or open a CSV file for writing
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ['Labels', 'Files']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    # Write the CSV header
    writer.writeheader()

    # Iterate through all 'hyperlink' folders under the 'year' folder
    for hyperlink_folder_name in os.listdir(year_folder_path):
        hyperlink_folder_path = os.path.join(year_folder_path, hyperlink_folder_name)

        # Check if the item in the 'year' folder is a directory (a 'hyperlink' folder)
        if os.path.isdir(hyperlink_folder_path):
            # Iterate through 'Complaint_denied' and 'Transfer' folders
            for folder_name in ['Complaint_denied', 'Transfer']:
                folder_path = os.path.join(hyperlink_folder_path, folder_name)

                # Check if the folder exists
                if os.path.exists(folder_path):
                    # List all files in the folder
                    files = os.listdir(folder_path)

                    # Iterate through the files in the folder
                    for file_name in files:
                        # Get the full path of the file
                        file_path = os.path.join(folder_path, file_name)

                        # Read the content of the file
                        with open(file_path, 'r', encoding='utf-8') as text_file:
                            file_content =text_file.read().replace('\n', ' ')
                        detected_language = detect(file_content)
                        if detected_language == 'en':
                        # Determine the label based on the folder name
                            label = 'Complaint denied' if folder_name == 'Complaint_denied' else 'Transfer'

                        # Write a row to the CSV file with the label and file content
                            writer.writerow({'Labels': label, 'Files': file_content})

print(f'Data from "{year_folder_path}" has been extracted and stored in "{csv_file_path}"')
```
